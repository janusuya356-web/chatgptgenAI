```python
import ollama
import gradio as gr

def chatbot(user_input):
    response = ollama.chat(
        model="phi3",
        messages=[
            {"role": "system", "content": "You are a helpful saving assistant."},
            {"role": "user", "content": user_input}
        ],
        options={
            "num_predict": 120
        }
    )

    # IMPORTANT: return ONLY STRING
    return response["message"]["content"]

demo = gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(lines=4, placeholder="Ask about saving money..."),
    outputs=gr.Textbox(label="Response"),
    title="My Saving Assistant"
)

demo.launch()
```
